# Visualisation of Data

```python
plt.figure(figsize=(20, 10))
df.groupby('Institution')['score'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('Institution')['course'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('Institution')['teaching'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('Institution')['feedback'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('Institution')['ratio'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('Institution')['spend'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('Institution')['tariff'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('Institution')['value'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('Institution')['career'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('Institution')['continuation'].mean().sort_values(ascending=False).plot(kind='bar')
```
