# Changing the Datatypes to float

```python
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

df['Followers'] = df['Followers'].apply(value_to_float)
```

```python
df.head()
```

```python
df['Authentic Engagement'] = df['Authentic Engagement'].apply(value_to_float)
```

```python
df.head()
```

```python
df['Engagement Avg.'] = df['Engagement Avg.'].apply(value_to_float)
```
