# Visualisation of Data

```python
plt.figure(figsize=(10, 10))
df.groupby('Country')['Percentage Score'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(10, 10))
top = df['Country'].value_counts().sort_values(ascending=False).head(8)
top.plot(kind='pie')
```

```python
Medal_win=df[df.Medal=="Yes"]
```

```python
Medal_win.head()
```

```python
Medal_win.tail()
```

```python
plt.figure(figsize=(10, 10))
top = Medal_win['Country'].value_counts().sort_values(ascending=False).head(8)
top.plot(kind='pie')
```

```python
plt.figure(figsize=(10, 10))
top = df['University'].value_counts().sort_values(ascending=False).head(8)
top.plot(kind='pie')
```

```python
df.University.value_counts().head(100)
```

# Visualisation of Data in India

```python
Coders_India=df[df.Country=="India"]
```

```python
Coders_India.head()
```

```python
plt.figure(figsize=(10, 10))
Coders_India.groupby('University')['num_of_top_100_last_20_year'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(10, 10))
top = Coders_India['University'].value_counts().sort_values(ascending=False).head(10)
top.plot(kind='pie')
```
